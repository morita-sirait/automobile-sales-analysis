```python
#import pandas library
import pandas as pd
```

```python
#read "Auto Sales data.csv"
df = pd.read_csv('Auto Sales data.csv')
```

```python
#show the several rows of data frame
df.head()
```

```python
#import seaborn library 
import seaborn as sns

#import malplotblib library
import matplotlib.pyplot as plt
```

```python
#show the order quantity of each day
sns.lineplot(x='ORDERDATE', y='QUANTITYORDERED', data=df)
plt.title('AUTOMOBILE SALES TRENDS OVER TIME')
plt.show()
```

```python
#show the distribution of order quantity
sns.histplot(df['QUANTITYORDERED'], kde=True)
plt.title('AUTOMOBILE SALES DISTRIBUTION')
plt.show()
```

```python
#show  order quantity based on category of automobile
sns.barplot(x='PRODUCTLINE', y='QUANTITYORDERED', data=df)
plt.title('SALES BY AUTOMOBILE CATEGORY')
plt.xticks(rotation=45)
plt.show()
```

```python
# compare the price distribution of product 
sns.boxplot(x='PRODUCTLINE', y='PRICEEACH', data=df)
plt.title('DISTRIBUTION BY PRICE OF AUTOMOBILE')
plt.xticks(rotation=90)
plt.show()
```

```python
#show the correalation between price and order quantity
sns.scatterplot(x='PRICEEACH', y='QUANTITYORDERED', data=df)
plt.title('THE RELATIONSHIP OF AUTOMOBILE PRICE AND ORDER QUANTITY')
plt.show()
```

```python
#show order quantity based on country
sns.barplot(x='COUNTRY', y='QUANTITYORDERED', data=df)
plt.title('SALES AUTOMOBILE BY REGION')
plt.xticks(rotation=90)
plt.show()
```
